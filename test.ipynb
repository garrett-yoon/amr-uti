```python
count = df_uti_features["demographics - is_white"].value_counts()
sns.set_style('whitegrid')
g = sns.barplot(x = count.index, y = count)
g.set(ylabel='Number of Observations')
_ = g.set(xticklabels=['Non-White', 'White'])
```

```python
count = df_uti_features.query('is_train == 1')["demographics - is_white"].value_counts()
sns.set_style('whitegrid')
g = sns.barplot(x = count.index, y = count)
g.set(ylabel='Number of Observations')
_ = g.set(xticklabels=['Non-White', 'White'])
```
